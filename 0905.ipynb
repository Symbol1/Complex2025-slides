## As vector field

```python
var('z')
```

```python
def V(f):
    def fr(u, v):
        zz = f(z = u + v * I)
        return real(zz)
    def fi(u, v):
        zz = f(z = u + v * I)
        return imag(zz)
    return (fr, fi)
```
